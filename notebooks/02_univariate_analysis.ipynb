# Univariate Analysis

## Objective
Analyze individual features to understand their distribution and patterns.

## Approach
- Use simple and business-friendly visualizations
- Separate numerical and categorical features
- Use generalized functions and loops

```python
#data handling
import pandas as pd
import numpy as np

#visualization
import matplotlib.pyplot as plt
```

```python
#loading dataset
df =pd.read_csv("../data/raw/automobile.csv", header=None, na_values="?") #the dataset has no column names and has missing values

#assigning names to the column
columns = [
    "symboling", "normalized_losses", "make", "fuel_type",
    "aspiration", "num_of_doors", "body_style", "drive_wheels",
    "engine_location", "wheel_base", "length", "width", "height",
    "curb_weight", "engine_type", "num_of_cylinders", "engine_size",
    "fuel_system", "bore", "stroke", "compression_ratio",
    "horsepower", "peak_rpm", "city_mpg", "highway_mpg", "price"
]
#Assigning column names to dataframe
df.columns = columns
```

```python
# Function to separate numerical and categorical columns
def get_column_types(df):
    
    # Select numerical columns
    numeric_cols = df.select_dtypes(include=np.number).columns.tolist()
    
    # Select categorical columns
    categorical_cols = df.select_dtypes(exclude=np.number).columns.tolist()
    
    return numeric_cols, categorical_cols


# Call function
num_cols, cat_cols = get_column_types(df)
```

```python
# Function to plot histogram for a column
def plot_histogram(df, column):
    
    plt.figure(figsize=(6,4))  # Create new figure
    
    # Plot histogram
    plt.hist(df[column].dropna(), bins=20,edgecolor='black')
    
    # Add title and labels
    plt.title(f"Distribution of {column}")
    plt.xlabel(column)
    plt.ylabel("Frequency")
    
    plt.show()
```

```python
# Apply histogram to all numerical columns
for col in num_cols:
    plot_histogram(df, col)
```

```python
# Function to plot bar chart for categorical columns
def plot_bar(df, column):
    
    plt.figure(figsize=(5,3))
    
    # Count values
    counts = df[column].value_counts()
    
    # Plot bar chart
    plt.bar(counts.index, counts.values)
    
    # Labels
    plt.title(f"Count of {column}")
    plt.xlabel(column)
    plt.ylabel("Count")
    
    plt.xticks(rotation=90)  # rotate labels if needed
    
    plt.show()
```

```python
for col in cat_cols:
    plot_bar(df, col)
```

 Univariate Analysis Summary

 Numerical Features

- Most numerical features such as wheel_base, length, width, and height show approximately normal distributions, indicating consistent vehicle dimensions.
- Features like engine_size, horsepower, and price are right-skewed, meaning most cars have lower values while a few cars have very high values (luxury or high-performance cars).
- Compression_ratio shows extreme skewness with a few unusually high values, indicating potential outliers.
- Curb_weight is moderately distributed, representing a mix of light, medium, and heavy vehicles.
- City_mpg and highway_mpg show slight right skewness, suggesting most cars have average fuel efficiency with few highly efficient vehicles.
- Peak_rpm is concentrated within a narrow range, indicating similar engine performance ranges across vehicles.
- Normalized_losses has missing values and a wide spread, showing inconsistency in insurance-related risk values.
- Price is heavily right-skewed, meaning most cars are affordable, with a few expensive cars acting as outliers.


Categorical Features

- Make: The dataset is dominated by brands like Toyota, Nissan, Mazda, and Honda, while luxury brands like Jaguar, Porsche, and Mercury have very few entries.
- Fuel Type: Majority of cars use gas, with very few diesel vehicles.
- Aspiration: Most cars are standard (naturally aspirated), while turbo cars are fewer.
- Number of Doors: Most cars have four doors, indicating family-oriented vehicles are more common.
- Body Style: Sedans and hatchbacks dominate the dataset, while convertibles and hardtops are rare.
- Drive Wheels: Most vehicles are front-wheel drive (FWD), followed by rear-wheel drive (RWD), with very few 4WD vehicles.
- Engine Location: Almost all cars have front-mounted engines, making it a highly imbalanced feature.
- Engine Type: OHC (Overhead Camshaft) engines dominate the dataset.
- Number of Cylinders: Most cars have four cylinders, indicating economy-focused vehicles.
- Fuel System: MPFI (Multi-Point Fuel Injection) is the most common system.



Key Insights

- Presence of skewness and outliers in features like price, engine_size, and compression_ratio.
- Dataset is dominated by economy and mid-range cars, with very few luxury vehicles.
- Several categorical features are highly imbalanced (e.g., engine_location, fuel_type).
- Clear separation exists between common vs rare categories, which is important for modeling.


Conclusion

Univariate analysis highlights the distribution, imbalance, and variability in the dataset. These insights will guide data cleaning, feature engineering, and model building in the next stages.
