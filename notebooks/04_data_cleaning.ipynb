```python
# Importing libraries

def import_lib():
    import pandas as pd
    import numpy as np
    return pd, np

pd, np = import_lib()
```

```python
# Load Dataset

def load_data(path):
    df = pd.read_csv(path, header=None)
    return df

df = load_data("../data/raw/automobile.csv")
```

```python
# Add column names (dataset has no headers)

def add_columns(df):
    columns = [
        "symboling", "normalized_losses", "make", "fuel_type",
        "aspiration", "num_of_doors", "body_style", "drive_wheels",
        "engine_location", "wheel_base", "length", "width", "height",
        "curb_weight", "engine_type", "num_of_cylinders", "engine_size",
        "fuel_system", "bore", "stroke", "compression_ratio",
        "horsepower", "peak_rpm", "city_mpg", "highway_mpg", "price"
    ]
    df.columns = columns  # assign column names
    return df

df = add_columns(df)
```

```python
# View first 5 rows of dataset

def show_head(df):
    return df.head()

show_head(df)
```

```python
# Check dataset structure and data types

def data_info(df):
    return df.info()

data_info(df)
```

```python
# Replace missing values marked as '?' with NaN

def replace_missing(df):
    import numpy as np
    df.replace("?", np.nan, inplace=True)
    return df

df = replace_missing(df)
```

```python
# Check missing values in each column

def missing_values(df):
    return df.isnull().sum()

missing_values(df)
```

```python
# Handle missing values (fill numeric with median, categorical with mode)

def fill_missing(df):
    for col in df.columns:
        if df[col].dtype == "object":
            df[col].fillna(df[col].mode()[0])
        else:
            df[col].fillna(df[col].median())
    return df

df = fill_missing(df)
```

```python
# Verify missing values after cleaning

def check_missing(df):
    return df.isnull().sum()

check_missing(df)
```

```python
# Convert numeric columns to proper data types

def convert_types(df):
    num_cols = [
        "normalized_losses", "wheel_base", "length", "width", "height",
        "curb_weight", "bore", "stroke", "compression_ratio",
        "horsepower", "peak_rpm", "city_mpg", "highway_mpg", "price"
    ]
    
    for col in num_cols:
        df[col] = df[col].astype(float)
    
    return df

df = convert_types(df)
```

```python
# Remove duplicate rows

def remove_duplicates(df):
    df = df.drop_duplicates()
    return df

df = remove_duplicates(df)
```

```python
# Clean categorical columns (lowercase + strip spaces)

def clean_categorical(df):
    cat_cols = df.select_dtypes(include="object").columns
    
    for col in cat_cols:
        df[col] = df[col].str.lower().str.strip()
    
    return df

df = clean_categorical(df)
```

```python
# Check final data types after cleaning

def check_dtypes(df):
    return df.dtypes

check_dtypes(df)
```

```python
# Final check after data cleaning

def final_check(df):
    print("Missing values check:\n", df.isnull().sum())
    print("\nData types:\n", df.dtypes)
    print("\nShape of dataset:", df.shape)

final_check(df)
```

```python
# Check dataset summary statistics

def summary_stats(df):
    return df.describe()

summary_stats(df)  #optional step
```

```python
# Save final cleaned dataset

import os

def save_data(df):
    os.makedirs("..\data/processed", exist_ok=True)
    df.to_csv("..\data/processed/cleaned_data.csv", index=False)

save_data(df)
```
