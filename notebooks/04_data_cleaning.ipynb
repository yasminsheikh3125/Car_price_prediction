```python
# Importing required libraries for data handling
import pandas as pd
import numpy as np
```

```python
# Loading dataset and assigning column names
# Also treating '?' as missing values during loading

columns = [
    "symboling", "normalized_losses", "make", "fuel_type",
    "aspiration", "num_of_doors", "body_style", "drive_wheels",
    "engine_location", "wheel_base", "length", "width", "height",
    "curb_weight", "engine_type", "num_of_cylinders", "engine_size",
    "fuel_system", "bore", "stroke", "compression_ratio",
    "horsepower", "peak_rpm", "city_mpg", "highway_mpg", "price"
]

df = pd.read_csv("../data/raw/automobile.csv", names=columns, na_values="?")

df.head()
```

```python
# Function to get overall understanding of dataset

def data_overview(df):
    print("Shape of dataset:", df.shape)
    
    print("\nDataset Info:")
    df.info()
    
    print("\nStatistical Summary:")
    print(df.describe())

# Call function
data_overview(df)
```

```python
# Checking number of missing values in each column

df.isnull().sum()
```

```python
# Filling missing values:
# * Numerical columns → median
# * Categorical columns → mode

for col in df.columns:
    
    if df[col].dtype == "object":
        df[col].fillna(df[col].mode()[0])
    
    else:
        df[col].fillna(df[col].median())

# Verify missing values handled
df.isnull().sum()
```

```python
# Converting necessary columns to numeric type

num_cols = [
    "normalized_losses", "wheel_base", "length", "width", "height",
    "curb_weight", "bore", "stroke", "compression_ratio",
    "horsepower", "peak_rpm", "city_mpg", "highway_mpg", "price"
]

for col in num_cols:
    df[col] = pd.to_numeric(df[col], errors='coerce')
```

```python
# Handle missing values again after type conversion

for col in df.columns:
    
    if df[col].dtype == "object":
        df[col].fillna(df[col].mode()[0])
    
    else:
        df[col].fillna(df[col].median())

# Check again
df.isnull().sum()
```

```python
# Remove duplicate rows if present

df.drop_duplicates(inplace=True)
```

```python
# Standardizing categorical values:
# convert to lowercase and remove extra spaces

for col in df.columns:
    
    if df[col].dtype == "object":
        df[col] = df[col].str.lower().str.strip()
```

```python
# Final check of dataset after cleaning

print("Final Shape:", df.shape)
print("\nFinal Missing Values:\n", df.isnull().sum())
print("\nFinal Data Types:\n", df.dtypes)
```

```python
# Saving cleaned dataset to processed folder

df.to_csv("../data/processed/cleaned_data.csv", index=False)
```
